```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
np.random.seed(45)
x1 = np.random.normal(0,1,(1000,1))
x2 = np.random.normal(0,1,(1000,1))
matcovar = np.diag((1.0,1.0))
matcovar[0,1] = 0.5
matcovar[1,0] = 0.5

print(matcovar)

chol = np.linalg.cholesky(matcovar)
print(chol)
```

```python
x = np.hstack((x1,x2))
nx = np.dot (x, chol)

print(nx)
print(np.shape(x1))
print(np.shape(nx[:,1]))

nxx = np.hstack((x1,nx[:,1].reshape(1000,1)))
```

```python

plt.figure(figsize=(5,5))
plt.scatter(nxx[:,0],nx[:,1])
```

```python
print(np.corrcoef(x1,nxx[:,1].reshape(1000,1),rowvar = 0 ))
```
